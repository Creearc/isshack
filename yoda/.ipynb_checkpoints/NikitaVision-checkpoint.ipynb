```python
def prep_ds(a):
    for i in range(0,len(a)):
        for j in range(0,10):
            a[i][1][j] = np.concatenate(a[i][1][j])

        a[i][1] = np.concatenate(a[i][1])
        a[i] = np.append(a[i][1],a[i][0])
    return(a) 
```

```python
arrs = prep_ds(arrs)
```

```python
np.array(arrs).shape
```

```python
def df_crt(a,b):
    nms = []
    for i in range(0,np.array(arrs).shape[1]-1):
        nms.append('X{}'.format(i))
    nms.append('sec')
    d = pd.DataFrame(data = a, columns=nms)
    d = pd.merge(d,b,'inner',['sec'])
    return(d)
```

```python
df = df_crt(arrs,sec)
```

```python
df['class'] = df['class'].astype('category')
```

```python
import seaborn as sns

sns.countplot(x = 'class', data=df)
```

```python
X = df.iloc[:, 0:-2]
y = df.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,stratify = y, random_state=13)
```

```python
y_train.unique()
```

```python
EPOCHS = 150
BATCH_SIZE = 512
LEARNING_RATE = 0.000001
```

```python
class LoadData(Dataset):
    
    def __init__(self, X_data, y_data):
        self.X_data = X_data
        self.y_data = y_data
        
    def __getitem__(self, index):
        return self.X_data[index], self.y_data[index]
        
    def __len__ (self):
        return len(self.X_data)


train_data = LoadData(torch.FloatTensor(np.array(X_train)), 
                       torch.FloatTensor(np.array(y_train)))
  
class TestData(Dataset):
    
    def __init__(self, X_data):
        self.X_data = X_data
        
    def __getitem__(self, index):
        return self.X_data[index]
        
    def __len__ (self):
        return len(self.X_data)
    

test_data = LoadData(torch.FloatTensor(np.array(X_test)),torch.FloatTensor(np.array(y_test)))
```

```python
train_loader = DataLoader(dataset=train_data, batch_size=BATCH_SIZE*40, shuffle=False)
test_loader = DataLoader(dataset=test_data, batch_size=BATCH_SIZE*40)
```

```python
class NClassifier(nn.Module):
    def __init__(self):
        super(NClassifier, self).__init__()
        self.layer_1 = nn.Linear(340, 680) 
        self.layer_3 = nn.Linear(680, 340)
        self.layer_5 = nn.Linear(340, 10)
        self.layer_out = nn.Linear(10, 1) 
        
        self.tanh = nn.Tanh()
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(p=0.005)
        self.batchnorm = nn.BatchNorm1d(680)
        self.sig = nn.Sigmoid()
        
    def forward(self, inputs):
        x = self.relu(self.layer_1(inputs))
        x = self.batchnorm(x)
        x = self.tanh(self.layer_3(x))
        x = self.tanh(self.layer_5(x))
        x = self.dropout(x)
        x = self.sig(self.layer_out(x))
        
        return x
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

print(device)
```

```python
model = NClassifier()
model.to(device)
print(model)
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
```

```python
from sklearn.metrics import f1_score

def score(y_p, y_t):
    y_pred_tag = torch.round(y_p)
    return f1_score(y_t.cpu().detach().numpy(),y_pred_tag.cpu().detach().numpy())
```

```python
model.train()
for e in range(1, EPOCHS+1):
    
    epoch_loss = 0
    epoch_acc = 0
    for X_batch, y_batch in train_loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)
        optimizer.zero_grad()
        
        y_pred = model(X_batch)
        
        loss = criterion(y_pred, y_batch.unsqueeze(1))
        acc = score(y_pred, y_batch.unsqueeze(1))
        
        loss.backward()
        optimizer.step()
        
        epoch_loss += loss.item()
        epoch_acc += acc.item()
        

    print(f'Epoch {e+0:03}: | Loss: {epoch_loss/len(train_loader):.5f} | F1: {epoch_acc/len(train_loader):.3f}')
```

```python
torch.save(model.state_dict(), 'saved_model.pth')
```
